### 실습1) 제품 입력후 상품명, 링크, 가격 추출하기

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import random
import pandas as pd

driver = webdriver.Chrome()
driver.get("https://www.naver.com")
time.sleep(random.uniform(0.15, 2.5))
driver.maximize_window()

#네이버 검색창에 클릭해서 커서 옮기기
search = driver.find_element(By.CSS_SELECTOR, '#query')
search.click()

#검색어 입력하기
search.send_keys('복숭아')
time.sleep(random.uniform(0.15, 2.5))
search.send_keys(Keys.ENTER)

morebtn = driver.find_element(By.CSS_SELECTOR, '.storeMoreLink-pc-module__link___OCNh8')
time.sleep(random.uniform(0.15, 2.5))
morebtn.click()
```

```python
print("current: ",driver.current_window_handle)
print("windows : ", driver.window_handles)
```

```python
driver.switch_to.window(driver.window_handles[1])
```

```python
# 전체 스크롤 하기
def scroll_down(driver):
    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(random.uniform(1, 2))
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height
```

```python
scroll_down(driver)

# 상품명, 링크, 가격 추출하기
items = driver.find_elements(By.CSS_SELECTOR,'.product_item__KQayS')
data = []
for item in items:
    name = item.find_element(By.CSS_SELECTOR, '.product_title__ljFM_>a').text
    link = item.find_element(By.CSS_SELECTOR, '.product_title__ljFM_>a').get_attribute('href')
    price = item.find_element(By.CSS_SELECTOR, '.price_num__Y66T7').text
    price = int(price.replace(',', '').replace('원', ''))
    data.append([name, link, price])
    print(name, link, price)

df=pd.DataFrame(data, columns=['제품명','링크','가격'])
df.to_excel('복숭아.xlsx',index=False)
```

### 실습2) 여러 페이지 추출하기

```python
# 셀레니움
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import random
import pandas as pd

# 전체 스크롤 하기
def scroll_down(driver):
    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(random.uniform(1, 2))
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height

driver = webdriver.Chrome()
driver.get("https://www.naver.com")
time.sleep(random.uniform(0.15, 2.5))
driver.maximize_window()

#네이버 검색창에 클릭해서 커서 옮기기
search = driver.find_element(By.CSS_SELECTOR, '#query')
search.click()

#검색어 입력하기
search.send_keys('복숭아')
time.sleep(random.uniform(0.15, 2.5))
search.send_keys(Keys.ENTER)

morebtn = driver.find_element(By.CSS_SELECTOR, '.storeMoreLink-pc-module__link___OCNh8')
time.sleep(random.uniform(0.15, 2.5))
morebtn.click()

driver.switch_to.window(driver.window_handles[1])

data = []

for page in range(1, 3):  
    print(f"크롤링 중: 페이지 {page}")

    scroll_down(driver)
    time.sleep(random.uniform(2, 4))

    # 상품명, 링크, 가격 추출하기
    items = driver.find_elements(By.CSS_SELECTOR,'.product_item__KQayS')
    for item in items:
        name = item.find_element(By.CSS_SELECTOR, '.product_title__ljFM_>a').text
        link = item.find_element(By.CSS_SELECTOR, '.product_title__ljFM_>a').get_attribute('href')
        price = item.find_element(By.CSS_SELECTOR, '.price_num__Y66T7').text
        price = int(price.replace(',', '').replace('원', ''))
        data.append([name, link, price])
        print(name, link, price)

    # 다음 페이지로 이동    
    next_button = driver.find_element(By.CSS_SELECTOR, 'a.pagination_next__kh_cw')
    next_button.click()
    time.sleep(random.uniform(2, 4))
    
driver.quit()

df = pd.DataFrame(data, columns=['제품명', '링크', '가격'])
df.to_excel('복숭아.xlsx', index=False)
```
